```python
from bs4 import BeautifulSoup
from pathlib import Path
import requests
import json
from tqdm import tqdm
```

```python
page = json.loads(BeautifulSoup(requests.get('https://wavendb.com/builds').text).find('div', attrs={'id':'app'}).attrs['data-page'])
Path('sample.json').write_text(json.dumps(page), encoding='utf-8')
```

```python
TYPE = "PVM"
VERSION = 2
URL = "https://wavendb.com/builds?page={page}&type={type}&weapon={weapon}&version={version}"
def get_builds(page, weapon, type=TYPE, version=VERSION):
    url = URL.format(page=page, type=type, weapon=weapon, version=version)
    return json.loads(BeautifulSoup(requests.get(url).text).find('div', attrs={'id':'app'}).attrs['data-page'])['props']['builds']['data']
```

```python
weapons = {p['id']:p['name_fr'] for p in page['props']['weapons']}
Path('weapons.json').write_text(json.dumps(weapons))
```

```python
gods = {p['id']:p['name_fr'] for p in page['props']['gods']}
Path('gods.json').write_text(json.dumps(weapons))
```

```python
data = []

for weapon in tqdm(weapons.keys()):
    data.extend(get_builds(1, weapon))
    data.extend(get_builds(2, weapon))

Path('data.json').write_text(json.dumps(data))
```
